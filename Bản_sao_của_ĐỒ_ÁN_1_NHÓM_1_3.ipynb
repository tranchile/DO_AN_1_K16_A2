Giao nhau SMA là một tín
hiệu giao dịch phổ biến khác. Khi SMA trong thời

```r
data_subset <- subset(data, Date >= as.Date("2023-12-13") & Date <= as.Date("2024-03-13"))
chartSeries(data_subset, theme = chartTheme('white'))
addSMA(n = 5, col = "blue")
addSMA(n = 25, col = "red")
```

Cả 2 đường SMA đều đi lên trong các thời kì cuối biểu thị xu huớng đi lên,đây có thể là tín hiệu tích cực về xu hướng tăng giá.

```r
data_subset <- subset(data, Date >= as.Date("2023-6-13") & Date <= as.Date("2023-9-13"))

# Vẽ biểu đồ giá cổ phiếu
chartSeries(data_subset, theme = chartTheme("white"))

# Vẽ đường MACD và đường tín hiệu
addMACD(fast=3, slow=8, signal=3,type="EMA")
```

Đường MACD: Đây là sự khác biệt giữa hai đường trung bình động (thường là trung bình động ngắn hạn và trung bình động dài hạn).
Đường tín hiệu (Signal line): Đây là một đường trung bình động của đường MACD


1.Giao nhau từ dưới lên (Crossover từ dưới lên):
Khi đường MACD (đường màu xanh) cắt lên trên đường tín hiệu (đường màu đỏ), đây thường là tín hiệu **MUA**.
Điều này cho thấy đà tăng của cổ phiếu đang gia tăng.


2.Giao nhau từ trên xuống (Crossover từ trên xuống):
Khi đường MACD cắt xuống dưới đường tín hiệu, đây thường là tín hiệu **BÁN**.
Điều này cho thấy đà giảm của cổ phiếu đang gia tăng.

```r
#Biểu đồ đường MACD
data_subset <- subset(data, Date >= as.Date("2023-9-13") & Date <= as.Date("2023-12-13"))
chartSeries(data_subset, theme = chartTheme('white'))
addMACD(fast=3, slow=8, signal=3,type="EMA")
```

Khi đường MACD và đường tín hiệu không cắt nhau và đang đi lên, điều này cho thấy sự gia tăng trong động lượng tăng giá.
Điều này có thể báo hiệu về sự tăng trưởng tiềm năng của cổ phiếu hoặc thị trường

```r
#Biểu đồ đường MACD
data_subset <- subset(data, Date >= as.Date("2023-12-13") & Date <= as.Date("2024-3-13"))
chartSeries(data_subset, theme = chartTheme('white'))
addMACD(fast=3, slow=8, signal=3,type="EMA")
```

Khi đường MACD và đường tín hiệu không cắt nhau và đang đi xuống, điều này cho thấy sự giảm động lượng giữa hai đường trung bình động.
Điều này có thể báo hiệu về sự "giảm giá" của cổ phiếu hoặc thị trường,có thể là thời điểm tốt để SELL.

BIỂU ĐỒ BAR CHART :Đối với dữ liệu khối lượng của giao dịch thỏa thuận và giao dịch khớp lệnh

```r
install.packages('tidyr')
library(tidyr)
```

```r
#Loại bỏ dấu phẩy nếu có và chuyển đổi sang numeric-dl số
data$NT_Volume = as.numeric(gsub(",", "", data$NT_Volume))
data$Trading_Volume = as.numeric(gsub(",", "", data$Trading_Volume))

#Tạo dataframe mới từ dữ liệu đã xử lý
data1=data.frame(data$Date, data$NT_Volume, data$Trading_Volume)
data1
colnames(data1)[1] = 'Date' #Đặt tên cột  đầu tiên của dataframe là Data
data1 = pivot_longer(data1, cols = c('data.NT_Volume','data.Trading_Volume'), names_to = 'Kind', values_to = 'Volume')
#Chuyển đổi cột Date thành thông tin quý
data1$Quarter = as.yearqtr(data1$Date)
#lấy danh sách các quý duy nhất
quarters = unique(data1$Quarter)
quarters
plots = list()

for (quarter in quarters) {
  #Lọc dữ liệu cho quý hiện tại
  quarter_data = subset(data1, Quarter == quarter)
  quarter_data
  #Định dạng tên quý theo "Quý Q YYYY"
  formatted_quarter = format.yearqtr(quarter, "Quý %q %Y")
  #Tạo biểu đồ bar chart cho từng quý

  p = ggplot() +
    geom_col(data = quarter_data, aes(x= Date, y= Volume, fill= Kind), position = 'dodge') +
    labs(title = paste('Trading Volume Chart for',formatted_quarter), x= 'Date', y= 'Trading Volume') +
    scale_fill_manual(values = c("darkgreen", "pink")) +
    theme_minimal() +
    theme(legend.position = "bottom", plot.title = element_text(face = "italic"))
  plots[[length(plots) + 1]] =p
}

#Sử dụng grid.arrange để sx các biểu đồ trên một trang
plot_grid = do.call(gridExtra :: grid.arrange, c(plots, nrow = 2, ncol = 2))
print(plot_grid)
```

Biểu đồ Box plot: Thể hiện phân phối của một biến và phát triển các ngoại lệ

```r
#Biến Net_Volume
#Loại bỏ dấu phẩy và chuyển đổi cột 'Net_Volume' sang numeric
data$Net_Volume = as.numeric(gsub(",", "", data$Net_Volume))
#Thêm cột quý
data$Quarter = as.yearqtr(data$Date)
#Tạo danh sách để lưu các biểu đồ
plots = list()
#tạo biểu đồ cho mỗi quý
unique_quarters = unique(data$Quarter)
for (quarter in unique_quarters) {
  #Lọc dữ liệu cho quý hiện tại
  quarter_data = subset(data, Quarter == quarter)
  #Định dạng tên quý theo "Quý Q YYYY"
  formatted_quarter = format.yearqtr(quarter, "Quý %q %Y")
  #vẽ bđ

  p = ggplot(quarter_data, aes(x = "", y = Net_Volume)) +
    geom_boxplot(fill = "lightblue", color = "darkred") +
    geom_point(data = quarter_data[quarter_data$Net_Volume > quantile(quarter_data$Net_Volume, 0.75) + 1.5 * IQR(quarter_data$Net_Volume),],  #Chọn điểm ngoại lệ
               aes(x = "", y = Net_Volume), color = "red", size = 3, shape = 19) +   #hiển thị điểm ngoại lệ
    labs(title = paste("Boxplot Chart of Net_Volume for", formatted_quarter), x = "", y = "Net Volume") +
    theme_minimal()
  theme(plot.title = element_text(hjust = 0.5))
  #Thêm biểu đồ vào danh sách
  plots[[length(plots) + 1]] = p
}

# Sắp xếp các biểu đồ trên một trang với bố cục c(2,2)
grid.arrange(grobs = plots, ncol = 2, nrow = 2)
```

```r
data= read.csv('/content/data.csv')
data$Change= as.numeric(gsub("\\(.*?\\)", "", data$Change))
data
data=data.frame(data$Price_Close,data$Price_Open,data$Price_High,row.names = data$Date)
print(data)
```

```r
library(NbClust)
library(cluster)
library(factoextra)
data1=data[150:209,]
data2=data[90:149,]
data3=data[30:89,]
data4=data[1:29,]
dtsc=function(i){
  i=scale(i)
  cl=get_clust_tendency(i, n=15, graph = TRUE)
  cl
  print(cl$plot)}
bieudo=function(i,a){
  km= kmeans(i, a, nstart = 25)
  fviz_cluster(km, i, geom = 'point', pointsize = 2, xlab = FALSE, ylab= FALSE, ggtheme = theme_minimal())}
check=function(i,a){
  km= kmeans(i, a, nstart = 25)
  sil= silhouette(km$cluster, dist(i))
  fviz_silhouette(sil)}
dtsc(data1)
bieudo(data1,3)
check(data1,3)
dtsc(data2)
bieudo(data2,3)
check(data2,3)
dtsc(data3)
bieudo(data3,3)
check(data3,3)
dtsc(data4)
bieudo(data4,3)
check(data4,3)
```

Lần lượt chúng ta sẽ có 3 biểu đồ đối với mỗi quý:
- Biểu đồ 1(hàm stsc): được sử dụng để ta có thể quan sát được nên tách dữ liệu làm mấy nhóm cho phù hợp với khối lượng và đồng thời biểu đồ vẽ ra có độ chính xác cao nhất
- Biểu đồ 2(hàm bieudo): đây sẽ là biểu đồ nhóm, sau khi quan sát thì ta có thể thấy các ngày được phân bố thành các nhóm theo phương sai
- Biểu đồ 3(hàm check): đây là biểu đồ cho thấy độ tin cậy của biểu đồ
