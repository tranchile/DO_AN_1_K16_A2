```r
library("readxl")
library("ggplot2")
library("dplyr")

file <- "/content/dulieu.xlsx"
data <- read_excel(file)
data
```

```r
# Tính tổng lượt nghe của mỗi năm
total_streams <- aggregate(in_spotify_playlists ~ Year, data, sum)
print(total_streams)

# Vẽ biểu đồ line cho xu hướng lượt nghe theo năm
ggplot(total_streams, aes(x = Year, y = in_spotify_playlists)) +
  geom_line() +
  geom_point() +
  labs(title = "Xu huong nghe theo nam",
       x = "Nam",
       y = "Tong so luot nghe") +
  theme_minimal() +
  scale_x_continuous(breaks = seq(min(total_streams$Year), max(total_streams$Year), by = 1))

```

```r
# Biểu đồ số lượng nghe trên nền tảng Spotify qua các năm
ggplot(data, aes(x = Song_name, y = streams, group = 1)) +
  geom_line(color = "blue") +
  geom_point(color = "blue", size = 0.1) +
  labs(title = "So luong nghe truc tuyen tren Spotify",
       x = "Ten bai hat",
       y = "So luong nguoi nghe") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1, vjust = 1, size = 8)) +
  theme(axis.title.x = element_text(size = 12))
```

```r
# Tính tổng lượt nghe của mỗi ca sĩ
Artist <- aggregate(streams ~ Artist, data, sum)

# Sắp xếp theo lượt nghe giảm dần và chọn ra top 5
top_Artists <- head(Artist[order(Artist$streams, decreasing = TRUE), ], 5)

# In ra top 5 ca sĩ có lượt nghe nhiều nhất
print("Top 5 ca sĩ có lượt nghe nhiều nhất:")
print(top_Artists)
#biểu đồ
ggplot(top_Artists, aes(x =Artist , y = streams)) +
  geom_col(fill = "skyblue", color = "black") +
  labs(title = "Top 5 Ca sĩ có lượt nghe nhiều nhất",
       x = "Ca sĩ",
       y = "Lượt nghe")
```

```r
# Tạo dữ liệu mẫu
data <- data.frame(
  Nen_Tang = c("Spotify", "Apple Music"),
  Playlist_Count = c("in_spotify_playlists", "in_apple_playlists") # Số lượng danh sách phát
)

# Vẽ biểu đồ cột so sánh giữa 2 nền tảng
ggplot(data, aes(x = Nen_Tang , y = Playlist_Count, fill = Nen_Tang)) +
  geom_bar(stat = "identity", width = 0.5) +
  labs(title = "So sanh so luong danh sach phat tren Spotify va Apple Music",
       x = "Nen Tang",
       y = "So luong danh sach phat") +
  theme_minimal()

```

```r
data_2019 <- subset(data, Year == 2019)

total_spotify <- sum(data_2019$in_spotify_playlists)
total_music <- sum(data_2019$in_apple_playlists)

barplot(
  c(total_spotify, total_music),
  names.arg = c("Spotify", "Music"),
  main = "So sanh so luong danh sach phat nam 2019",
  xlab = "Nen Tang",
  ylab = "So luong bai hat",
  col = c("blue", "red"),
  ylim = c(0, max(total_spotify, total_music) * 1.1),
  beside = TRUE
)


```

```r
data_2023 <- subset(data, Year == 2023)

total_spotify <- sum(data_2023$in_spotify_playlists)
total_music <- sum(data_2023$in_apple_playlists)

barplot(
  c(total_spotify, total_music),
  names.arg = c("Spotify", "Music"),
  main = "So sanh so luong danh sach phat nam 2023",
  xlab = "Nen Tang",
  ylab = "So luong bai hat",
  col = c("blue", "red"),
  ylim = c(0, max(total_spotify, total_music) * 1.1),
  beside = TRUE
)
```

```r
# Tính tổng số bài hát cho mỗi năm
total_spotify <- c(sum(data_2019$in_spotify_playlists),
                   sum(data_2020$in_spotify_playlists),
                   sum(data_2021$in_spotify_playlists),
                   sum(data_2022$in_spotify_playlists),
                   sum(data_2023$in_spotify_playlists))

total_apple_music <- c(sum(data_2019$in_apple_playlists),
                        sum(data_2020$in_apple_playlists),
                        sum(data_2021$in_apple_playlists),
                        sum(data_2022$in_apple_playlists),
                        sum(data_2023$in_apple_playlists))

# Hiển thị tổng số bài hát
print("Tổng số bài hát của các năm của nền tảng Spotify(2019-2023):")
print(total_spotify)
print("Tổng số bài hát của các năm của nền tảng Apple Music(2019-2023):")
print(total_apple_music)

# Tạo dữ liệu cho biểu đồ
data_total <- data.frame(
  Year = c(2019, 2020, 2021, 2022, 2023),
  Spotify = total_spotify,
  Apple_Music = total_apple_music
)

library(reshape2)
data_long_total <- melt(data_total, id.vars = "Year", variable.name = "Platform", value.name = "Total_Song_Count")

# Vẽ biểu đồ cột
bar_plot_total <- ggplot(data_long_total, aes(x = Year, y = Total_Song_Count, fill = Platform)) +
  geom_bar(stat = "identity", position = "dodge", width = 0.5) +
  labs(title = "So sánh số lượng bài hát tham gia của Spotify và Apple Music (2019-2023)",
       x = "Năm",
       y = "Tổng số bài hát",
       fill = "Nền tảng") +
  theme_minimal()
bar_plot_total <- bar_plot_total +
  scale_x_continuous(breaks = seq(min(data_total$Year), max(data_total$Year), by = 1)) +
  scale_y_continuous(labels = scales::comma)

# Hiển thị biểu đồ cột
print(bar_plot_total)
```

```r
# Phân tích hồi quy cho in_spotify_playlists và in_spotify_charts
spotify_model <- lm(in_spotify_charts ~ in_spotify_playlists, data=data)
summary(spotify_model) # In kết quả tóm tắt của mô hình

# Kiểm tra giả định của mô hình
par(mfrow=c(2,2))
plot(spotify_model)

```

```r
# Trực quan hóa kết quả của Spotify
ggplot(data, aes(x=in_spotify_playlists, y=in_spotify_charts)) +
  geom_point(color="red") +
  geom_smooth(method="lm", se=FALSE, color="blue") +
  labs(x="So luong playlist tren Spotify", y="So luong bai hat tren Spotify",
       title="Phân tích hoi quy Spotify",
       caption="Đuong hoi quy: mau xanh, Điem du lieu: mau đo")

```

```r
# Phân tích hồi quy cho in_apple_playlists và in_apple_charts
apple_model <- lm(in_apple_charts ~ in_apple_playlists, data=data)
summary(apple_model) # In kết quả tóm tắt của mô hình

# Kiểm tra giả định của mô hình
par(mfrow=c(2,2))
plot(apple_model)
```

```r
# # Trực quan hóa kết quả của Music
ggplot(data, aes(x=in_apple_playlists, y=in_apple_charts)) +
  geom_point(color="red") +
  geom_smooth(method="lm", se=FALSE, color="blue") +
  labs(x="So luong playlist tren Apple Music", y="So luong bai hat tren Apple Music",
       title="Phan tich hoi quy Apple Music",
       caption="Đuong hoi quy: mau xanh, Điem du lieu: mau đo")
```

```r
#Hệ số tương quan giữa 2 biến in_spotify_playlists và streams

# Chọn hai cột dữ liệu
column1 <- data$in_spotify_playlists
column2 <- data$streams
correlation_matrix <- cor(column1,column2)
# Tính độ đồng biến
covariance <- cov(column1, column2)

print("Hệ số tương quan giữa 2 biến in_spotify_playlists và streams")
print(correlation_matrix)
print(paste("Độ đồng biến:", covariance))

# Vẽ biểu đồ scatter plot
ggplot(data, aes(x = column1, y = column2)) +
  geom_point() +
  labs(title = "Scatter Plot giua in_spotify_playlists và streams",
       x = "in_spotify_playlists",
       y = "streams") +
  theme_minimal()

```

```r
#hệ số tương quan giữa 2 biến in_spotify_playlists và in_apple_playlists
# Chọn hai cột dữ liệu
column1 <- data$in_spotify_playlists
column3 <- data$in_apple_playlists
correlation_matrix <- cor(column1,column3)
# Tính độ đồng biến
covariance <- cov(column1, column3)

print("Hệ số tương quan giữa 2 biến in_spotify_playlists và in_apple_playlists")
print(correlation_matrix)
print(paste("Độ đồng biến:", covariance))

# Vẽ biểu đồ scatter plot
ggplot(data, aes(x = column1, y = column3)) +
  geom_point() +
  labs(title = "Scatter Plot giua in_spotify_playlists và in_apple_playlists",
       x = "in_spotify_playlists",
       y = "in_apple_playlists") +
  theme_minimal()
```

```r
#Hệ số tương quan giữa 2 biến in_apple_playlists và streams
correlation_matrix <- cor(column3,column2)
# Tính độ đồng biến
covariance <- cov(column3, column2)

print("Hệ số tương quan giữa 2 biến in_apple_playlists và streams")
print(correlation_matrix)
print(paste("Độ đồng biến:", covariance))

# Vẽ biểu đồ scatter plot
ggplot(data, aes(x = column3, y = column2)) +
  geom_point() +
  labs(title = "Scatter Plot giua in_apple_playlists và streams",
       x = "in_apple_playlists",
       y = "streams") +
  theme_minimal()

```

```r
# Chuẩn hóa dữ liệu
numeric_data <- data[, sapply(data, is.numeric)]

normalized_data <- scale(numeric_data)
print(numeric_data)
```

```r
# Tính toán ma trận hiệp phương sai
covariance_matrix <- cov(normalized_data)
```

```r
# Tìm vectơ riêng và giá trị riêng
eigenvalues <- eigen(covariance_matrix)$values
eigenvectors <- eigen(covariance_matrix)$vectors
```

```r
# Lựa chọn số lượng thành phần chính
total_variance <- sum(eigenvalues)
explained_variance <- eigenvalues / total_variance
cumulative_variance <- cumsum(explained_variance)

# Chọn số lượng thành phần chính
num_components <- which.max(cumulative_variance >= 0.9)  # Chọn số lượng có thể giải thích 90% phương sai
```

```r
# Biến đổi dữ liệu sang không gian thành phần chính
pca_data <- as.matrix(normalized_data) %*% eigenvectors[, 1:num_components]
```

```r
# Biểu đồ thanh cho các thành phần chính
bar_data <- data.frame(tp_chinh = paste0("PC", 1:num_components),
                       giai_tich_psai = explained_variance[1:num_components])

bar_plot <- ggplot(bar_data, aes(x = tp_chinh, y = giai_tich_psai)) +
  geom_bar(stat = "identity", fill = "skyblue", width = 0.5) +
  labs(title = "thanh phan chinh và giai tich phuong sai",
       x = "thanh phan chinh",
       y = "giai tich phuong sai") +
  theme_minimal()

print(bar_plot)
```
