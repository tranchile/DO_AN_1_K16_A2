
**1.6 LÀM SẠCH CỘT DỮ LIỆU CỘT BLOOD PRESSURE.**




Các mức độ của huyết áp thường được phân loại như sau

- Normal (Bình thường): Huyết áp tâm trương dưới 120 mmHg và huyết áp tâm thu dưới 80 mmHg.
- Prehypertension (Tiền cao huyết áp): Huyết áp tâm trương từ 120 đến 139 mmHg hoặc huyết áp tâm thu từ 80 đến 89 mmHg.
- Stage high blood pressure 1: Huyết áp tâm trương từ 140 đến 159 mmHg hoặc huyết áp tâm thu từ 90 đến 99 mmHg.
- Stage high blood pressure 2: Huyết áp tâm trương 160 mmHg trở lên hoặc huyết áp tâm thu 100 mmHg trở lên.


```r
column_name <- "Blood Pressure"
# Đọc dữ liệu từ cột Blood Pressure
column_data <- data[[column_name]]
# Hiển thị dữ liệu cột đã chọn
print(column_data)
huyet_ap_split <- strsplit(column_data, "/")
# Lấy huyết áp tâm trương và huyết áp tâm thu thành các vector riêng biệt
huyet_ap_truong <- as.numeric(sapply(huyet_ap_split, "[", 1))
huyet_ap_thu <- as.numeric(sapply(huyet_ap_split, "[", 2))

# Phân loại huyết áp
Blood_calssify_huyet_ap <- function(huyet_ap_truong, huyet_ap_thu) {
  if (huyet_ap_truong < 120 & huyet_ap_thu < 80) {
    return("Normal")
  } else if (huyet_ap_truong >= 120 & huyet_ap_truong <= 139 | huyet_ap_thu >= 80 & huyet_ap_thu <= 89) {
    return("Prehypertension")
  } else if (huyet_ap_truong >= 140 & huyet_ap_truong <= 159 | huyet_ap_thu >= 90 & huyet_ap_thu <= 99) {
    return("Stage high blood pressure 1")
  } else if (huyet_ap_truong >= 160 | huyet_ap_thu >= 100) {
    return("Stage high blood pressure 2")
  }

}
# Áp dụng hàm phân loại cho từng cặp huyết áp
Blood_calssify <- mapply(Blood_calssify_huyet_ap, huyet_ap_truong, huyet_ap_thu)
data[[column_name]]<- Blood_calssify
print(Blood_calssify)
```

**1.7 PHÂN CHIA LẠI MỨC ĐỘ NHỊP TIM.**



- Thấp(Low) : Là nhịp tim dưới 65 nhịp/phút
- Bình thường(Normal) : Là nhịp tim từ 65 -> 85 nhịp/phút
- Cao(High) : Là nhịp tim trên 85 nhịp/phút

```r
column_names <- "Heart Rate"
column_dataH <- data[[column_names]]
print(column_dataH)
classify_heart_rate <- function(heart_rate) {
  if (heart_rate <= 65) {
    return("Low")
  } else if (heart_rate < 85) {
    return("Normal")
  } else {
    return("High")
  }
}
Heart_classify <- sapply(column_dataH, classify_heart_rate)
print(Heart_classify)
```

**3.2 HỆ SỐ TƯƠNG QUAN.**


 Hệ số tương quan là một đại lượng số học được sử dụng để `đo lường mức độ mối quan hệ giữa hai biến số`. Nó cho biết đến mức độ nào các biến số di chuyển cùng nhau trong một tập dữ liệu. Hệ số tương quan có giá trị từ -1 đến 1, trong đó:

 - Nếu hệ số tương quan gần bằng 1, có một mối tương quan tuyến tính mạnh dương, tức là khi một biến tăng thì biến kia cũng tăng theo cùng hướng.

 - Nếu hệ số tương quan gần bằng -1, có một mối tương quan tuyến tính mạnh âm, tức là khi một biến tăng thì biến kia giảm theo cùng hướng.

 - Nếu hệ số tương quan gần bằng 0, không có mối tương quan tuyến tính giữa hai biến.

HỆ SỐ TƯƠNG QUAN GIỮA BIẾN 'QUALITY OF SLEEP' VỚI TẤT CẢ CÁC BIẾN SỐ CÒN LẠI

3. **Độ đồng biến (Covariance)**: Độ đồng biến giữa "Mức độ stress" và "Chất lượng giấc ngủ" là **-1.213**, một giá trị âm. Điều này cũng cho thấy một mối quan hệ đồng biến âm giữa hai biến, tức là khi một biến tăng, biến kia giảm.

```r
# Biến đổi cột Sleep Disorder sang tiếng anh
colnames('Sleep Disorder')
data[data=="Không"]<-"No Problem"
data[data=="Chứng ngưng thở khi ngủ"]<-"Sleep Apnea"
data[data=="Mất ngủ"]<-"Insomnia"
print(data["Sleep Disorder"])
```

```r

# Tính tổng số lượng của mỗi triệu chứng khi ngủ
total_counts <- data %>%
  group_by(`Sleep Disorder`) %>%
  summarise(total = n())

# Tính số lượng mỗi triệu chứng khi ngủ theo từng ngành nghề
Sleep_Disorder_counts <- data %>%
  group_by(Occupation, `Sleep Disorder`) %>%
  summarise(count = n(), .groups = "drop") %>%
  ungroup() %>%
  left_join(total_counts, by = "Sleep Disorder") %>%
  mutate(percent = count / total * 100)
Sleep_Disorder_counts

# Tạo bảng màu
symptom_colors <- c("No Problem" = "orange", "Sleep Apnea" = "pink", "Insomnia" = "yellow")

# Vẽ biểu đồ tròn
ggplot(Sleep_Disorder_counts, aes(x = "", y = percent, fill = `Sleep Disorder`, label = paste0(round(percent), "%"))) +
  geom_bar(width = 1, stat = "identity", color = "black") +  # Thêm viền cho các cột
  coord_polar(theta = "y", start = 0) +  # Biểu đồ tròn
  facet_wrap(~Occupation) +  # Phân chia biểu đồ theo từng ngành nghề
  scale_fill_manual(values = symptom_colors) +  # Đặt màu sắc cho các phần
  theme(legend.position = "bottom") +  # Đặt chú thích ở dưới
  labs(title = "      PIE CHART OF SLEEP SYMPTOMS BY OCCUPATION",
       fill = "SLEEP SYMPTOMS :",
       x = NULL, y = NULL)  # Ẩn nhãn trục x và y
```

- Insomnia là một tình trạng khó ngủ, khiến người bệnh gặp khó khăn trong việc đầu óc và cơ thể đều không thể nghỉ ngơi đủ trong khi ngủ. Insomnia không chỉ đơn thuần là khó ngủ mà còn liên quan đến việc thức dậy sớm, thức dậy nhiều lần trong đêm, hoặc cảm giác không hài lòng với giấc ngủ đã có.

  - Ngành nghề có tỷ lệ người gặp triệu chứng giấc ngủ cao nhất là kế toán, với 60% người gặp mất ngủ và 40% người gặp ngưng thở khi ngủ.

```r
# Loại bỏ cột "ID" từ dữ liệu
data <- data[, !names(data) %in% "ID"]
# Chuyển đổi các biến phân loại thành factor để sử dụng trong mô hình cây quyết định.
# Factor giúp R hiểu rằng các giá trị của biến này chỉ thuộc vào một tập hợp hữu hạn các giá trị có thể, và không phải là các giá trị số liên tục.
data$Gender <- as.factor(data$Gender)
data$Age <- as.factor(data$Age)
data$Occupation <- as.factor(data$Occupation)
data$Sleep_Duration <- as.factor(data$`Sleep Duration`)
data$Quality_of_Sleep <- as.factor(data$`Quality of Sleep`)
data$Physical_Activity_Level <- as.factor(data$`Physical Activity Level`)
data$Stress_Level <- as.factor(data$`Stress Level`)
data$Blood_Pressure <- as.factor(data$`Blood Pressure`)
data$BMI <- as.factor(data$BMI)
data$Heart_Rate <- as.factor(data$`Heart Rate`)
data$Daily_Steps <- as.factor(data$`Daily Steps`)
data$Sleep_Disorder <- as.factor(data$`Sleep Disorder`)
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra.
set.seed(123)
train_index <- sample(1:nrow(data), 0.7 * nrow(data)) # 'sample' để chia dữ liệu thành tập huấn luyện (train_data) và tập kiểm tra (test_data) theo tỷ lệ 70:30.
train_data <- data[train_index, ]
test_data <- data[-train_index, ]

# Xây dựng mô hình cây quyết định
tree_model <- rpart(Quality_of_Sleep ~ ., data = train_data, method = "class",
                    control = rpart.control(minsplit = 10, xval = 0))

# biến phụ thuộc là Quality_of_Sleep, và chúng ta sử dụng tất cả các biến khác làm biến dự đoán. Cách xây dựng mô hình được xác định bởi method = "class" để tạo một mô hình phân loại.
# Dự đoán
predictions <- predict(tree_model, newdata = test_data, type = "class")
print("Dự đoán chất lượng giấc ngủ của mô hình test_data là : ")
predictions
# Đánh giá mô hình
accuracy <- mean(predictions == test_data$Quality_of_Sleep)
print(paste("Đánh Giá Mô Hình Có Độ Chính Xác Là :", accuracy))


```

**Đánh giá hiệu suất của mô hình bằng confusion matrix**

- Ma trận nhầm lẫn (confusion matrix) là một công cụ quan trọng trong việc đánh giá hiệu suất của một mô hình dự đoán trong machine learning và thống kê. Nó thể hiện sự so sánh giữa dự đoán của mô hình và giá trị thực tế của dữ liệu.

- Ma trận nhầm lẫn thường được sử dụng trong các bài toán phân loại (classification), nơi mà mô hình cố gắng dự đoán các nhãn hoặc lớp cho dữ liệu đầu vào. Ma trận này có dạng một bảng, trong đó các hàng thể hiện các nhãn thực tế và các cột thể hiện các nhãn được dự đoán bởi mô hình.

```r
                   Predicted Class
               |  Class 1  |  Class 2  |  ...  |  Class n  |
------------------------------------------------------------
Actual Class 1 |    TP1    |    FP1    |  ...  |    FN1    |
------------------------------------------------------------
Actual Class 2 |    FP2    |    TP2    |  ...  |    FN2    |
------------------------------------------------------------
      ...      |    ...    |    ...    |  ...  |    ...    |
------------------------------------------------------------
Actual Class n |    FPn    |    TPn    |  ...  |    FNn    |
```

- Trong đó:

  - TP (True Positive): Số lượng điểm dữ liệu thuộc lớp thực tế và được mô hình dự đoán chính xác.
  - FP (False Positive): Số lượng điểm dữ liệu không thuộc lớp thực tế nhưng lại được mô hình dự đoán là thuộc lớp đó (lỗi dự đoán sai).
  - FN (False Negative): Số lượng điểm dữ liệu thuộc lớp thực tế nhưng lại không được mô hình dự đoán là thuộc lớp đó (lỗi dự đoán sai).
  - TN (True Negative): Số lượng điểm dữ liệu không thuộc lớp thực tế và mô hình dự đoán đúng là không thuộc lớp đó.

```r

# Đánh giá hiệu suất của mô hình bằng confusion matrix
confusion_matrix <- table(predicted = predictions, actual = test_data$Quality_of_Sleep)
print(confusion_matrix)
```

- Ma trận nhầm lẫn cung cấp cái nhìn tổng quan về hiệu suất của mô hình và có thể sử dụng để tính toán các chỉ số đánh giá như độ chính xác (accuracy), độ nhạy (recall), độ chính xác của dự đoán tích cực (precision), và F1-score.

- Các chỉ số đánh giá như độ chính xác (accuracy), độ nhạy (recall), độ chính xác của dự đoán tích cực (precision) và F1-score là các độ đo được sử dụng để đánh giá hiệu suất của mô hình phân loại. Dưới đây là ý nghĩa của từng chỉ số:

  - Độ chính xác (Accuracy): Độ chính xác là tỷ lệ giữa số lượng dự đoán đúng (true positives và true negatives) trên tổng số quan sát trong tập dữ liệu.

  - Độ nhạy (Recall): Độ nhạy đo lường khả năng của mô hình trong việc phát hiện tất cả các trường hợp thuộc lớp tích cực (positive class). Nó là tỷ lệ giữa số lượng dự đoán đúng thuộc lớp tích cực (true positives) trên tổng số quan sát thực sự thuộc lớp tích cực.

  - Độ chính xác của dự đoán tích cực (Precision): Độ chính xác của dự đoán tích cực đo lường tỷ lệ của các trường hợp dự đoán là tích cực mà thực sự là tích cực. Nó là tỷ lệ giữa số lượng dự đoán đúng tích cực (true positives) trên tổng số lượng dự đoán tích cực (true positives và false positives).

  - F1-score: F1-score là trung bình điều hòa giữa độ nhạy và độ chính xác của dự đoán tích cực. Nó cung cấp một đánh giá tổng thể về hiệu suất của mô hình phân loại, kết hợp cả khả năng phát hiện các trường hợp tích cực và độ chính xác của dự đoán.

=▶ Tất cả các chỉ số này được sử dụng để đánh giá mức độ chính xác và độ chính xác của mô hình phân loại trên dữ liệu thử nghiệm.


```r
# Tính các chỉ số đánh giá từ ma trận nhầm lẫn
calculate_metrics <- function(conf_matrix) {
  TP <- diag(conf_matrix)  # Số lượng dự đoán đúng positive
  TN <- sum(conf_matrix) - rowSums(conf_matrix) - colSums(conf_matrix) + diag(conf_matrix)  # Số lượng dự đoán đúng negative
  FP <- colSums(conf_matrix) - TP  # Số lượng dự đoán sai positive
  FN <- rowSums(conf_matrix) - TP  # Số lượng dự đoán sai negative

# Độ chính xác (Accuracy)
  accuracy <- sum(TP) / sum(conf_matrix)

# Độ nhạy (Recall)
  recall <- TP / (TP + FN)

# Độ chính xác của dự đoán tích cực (Precision)
  precision <- TP / (TP + FP)

# F1-score
  f1_score <- 2 * precision * recall / (precision + recall)

# Tạo dataframe chứa các chỉ số đánh giá
  metrics <- data.frame(
    Accuracy = accuracy,
    Recall = recall,
    Precision = precision,
    F1_Score = f1_score
  )

  return(metrics)
}

# Sử dụng hàm calculate_metrics để tính toán các chỉ số đánh giá từ ma trận nhầm lẫn
conf_matrix <- confusion_matrix
metrics <- calculate_metrics(conf_matrix)
print(metrics)
```

```r
data_new<-data
BMI<-as.numeric(factor(data_new$BMI, levels=c("Normal", "Obesity","Pre-obesity (Overweight)"), labels=c(0,1,2)))
Pressure<-as.numeric(factor(data_new$`Blood Pressure`, levels=c("Normal", "Stage high blood pressure 1","Prehypertension"), labels=c(0,1,2)))
Disorder<-as.numeric(factor(data_new$`Sleep Disorder`, levels=c("No Problem", "Insomnia","Sleep Apnea"), labels=c(0,1,2)))
```

 **MỐI QUAN HỆ QUALITY OF SLEEP VÀ CÁC BIẾN THÔNG QUA HỒI QUY ĐA BIẾN**

```r
Quality<-data_new$`Quality of Sleep`
Duration<- data_new$`Sleep Duration`
Activity<-data_new$`Physical Activity Level`
Stress<-data_new$`Stress Level`
Heart<-data_new$`Heart Rate`
Daily<-data_new$`Daily Steps`
```

```r
data_lm<-lm(Quality ~ Duration+ Activity +Stress +BMI+Pressure+Heart+Daily+ Disorder)
summary(data_lm)
```

* Intercept(Hệ số góc): Intercept là 11.51. Điều này có nghĩa là khi tất cả các biến độc lập đều bằng 0,Quality of Sleep được dự đoán là 11.51. Tuy nhiên,với số liệu trên,nó chỉ là một giá trị tham khảo vì không có biến độc lập nào có thể đạt được giá trị 0.

* Mô hình này có R-squared là 0.8976, tức là khoảng 89.76% sự biến thiên của chất lượng giấc ngủ được giải thích bởi các biến độc lập trong mô hình.
