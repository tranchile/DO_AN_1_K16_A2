4. Phân tích mối liên hệ giữa các biến:
1.    Biểu đồ (scatter plot) và ma trận tương quan (correlation matrix) được sử dụng để phân tích mối quan hệ giữa các biến.
2.    Biểu đồ (scatter plot):
●	•   	Biểu đồ scatter plot thể hiện mối quan hệ giữa hai biến thông qua điểm dữ liệu trên mặt phẳng.
●	•   	Nó cho phép quan sát mức độ phân tán của dữ liệu và xác định xem liệu có một mối quan hệ     tương quan giữa các biến hay không.
3.    Ma trận tương quan (correlation matrix):
●	Ma trận tương quan thể hiện mối quan hệ tương quan giữa tất cả các cặp biến trong tập dữ liệu.
●	Nó cung cấp thông tin về hướng (dương hoặc âm) và độ mạnh của mối quan hệ giữa các biến.
 4.  Hệ số tương quan (correlation coefficient) và độ đồng biến (covariance):
●	Hệ số tương quan là một độ đo chuẩn hóa của mối quan hệ tương quan giữa hai biến.
●	Nó thường được biểu diễn bằng giá trị từ -1 đến 1, trong đó 1 thể hiện mối quan hệ tương quan hoàn toàn dương, -1 thể hiện mối quan hệ tương quan hoàn toàn âm, và 0 thể hiện không có mối quan hệ tương quan.
Độ đồng biến (covariance) đo lường hướng của mối quan hệ giữa hai biến. Giá trị dương cho biết các biến tăng cùng nhau, trong khi giá trị âm cho biết chúng có xu hướng tăng ngược nhau

# Ma trận tương quan


```r
library(corrplot)
library(RColorBrewer)
load("Women_2022.rda")
View(Women_2022)
head(Women_2022)
summary(Women_2022)
D=Women_2022[3:14]
M <- cor(D, method = "spearman")
corrplot(M, method = "pie")
corrplot(M, type = "upper")
corrplot(M, type = "upper", order = "hclust")
col <- colorRampPalette(c("red", "blue", "violet"))(3)
corrplot(M, type = "upper", order = "hclust", col = col)
corrplot(M, type = "upper", order = "hclust", col = c("black", "white"), bg = "pink")
corrplot(M, type = "upper", order = "hclust", tl.col = "black", tl.srt = 90)
cor.mtest <- function(mat, ...) {
  mat <- as.matrix(mat)
  n <- ncol(mat)
  p.mat <- matrix(NA, n, n)
  diag(p.mat) <- 0
  for(i in 1:(n-1)) {
    for(j in (i+1):n) {
      tmp <- cor.test(mat[,i], mat[,j], ...)
      p.mat[i,j] <- p.mat[j,i] <- tmp$p.value
    }
  }
  colnames(p.mat) <- rownames(p.mat) <- colnames(mat)
  p.mat
}
p.mat <- cor.mtest(D)
corrplot(M, type = "upper", order = "hclust", p.mat = p.mat, sig.level = 0.01, insig = "blank")

my_palette <- colorRampPalette(c("#BB4444", "#EE9988", "#FFFFFF", "#77AADD", "#4477AA"))
corrplot(M, method = "color", col = my_palette(200), type = "upper",
         order = "hclust", addCoef.col = "black", tl.col = "black",
         tl.srt = 45, p.mat = p.mat, sig.level = 0.01, insig = "blank",
         diag = FALSE)

```

#6 : Một số chủ đề nâng cao

#Lựa chọn một trong các chủ để nâng cao sau để phân tích dữ liệu:

    
    1.Phân tích thành phần chính (PCA)
    2.Phân tích nhóm (cluster analyst)
    3.Khai phá dữ liệu (data mining)

Trong bối cảnh phân tích dữ liệu hiện đại, việc chọn phương pháp phân tích phù hợp là vô cùng quan trọng để hiểu rõ và khai thác hiệu quả thông tin từ dữ liệu. Dữ liệu về các chỉ số dân số và xã hội của phụ nữ tại các tỉnh thành Việt Nam năm 2022 cung cấp một lượng thông tin phong phú nhưng cũng rất phức tạp. Với những đặc điểm như số lượng biến số lớn và mối quan hệ phức tạp giữa các biến, việc lựa chọn phương pháp phân tích phải đảm bảo khả năng giảm thiểu độ phức tạp của dữ liệu, đồng thời vẫn giữ lại được các thông tin quan trọng.

Phân tích Thành phần Chính (Principal Component Analysis - PCA) nổi bật lên như một phương pháp lý tưởng trong tình huống này. Dưới đây là những lý do chính vì sao chọn PCA trong ba chủ đề nâng cao được đề xuất:

1. **Giảm thiểu chiều dữ liệu**: Dữ liệu với nhiều biến số có thể gây khó khăn trong việc phân tích và trực quan hóa. PCA giúp giảm số chiều của dữ liệu mà vẫn giữ được phần lớn thông tin, giúp cho việc phân tích trở nên dễ dàng và hiệu quả hơn.

2. **Tìm ra các mẫu ẩn và mối quan hệ**: PCA có khả năng tìm ra các mẫu ẩn và mối quan hệ giữa các biến số thông qua các thành phần chính. Điều này giúp chúng ta hiểu rõ hơn về cấu trúc nội tại của dữ liệu và các yếu tố ảnh hưởng chính.

3. **Tăng hiệu quả tính toán**: Việc giảm số chiều dữ liệu không chỉ giúp cho việc trực quan hóa mà còn tăng hiệu quả tính toán, đặc biệt khi xử lý các mô hình phức tạp và dữ liệu lớn.

4. **Trực quan hóa dữ liệu**: PCA giúp tạo ra các biểu đồ trực quan, như biểu đồ scatterplot của các thành phần chính, giúp dễ dàng nhận diện các mẫu và xu hướng trong dữ liệu.

5. **Chuẩn bị dữ liệu cho các phân tích khác**: PCA không chỉ là một phương pháp phân tích độc lập mà còn là bước chuẩn bị dữ liệu quan trọng cho các phân tích khác như phân tích nhóm (cluster analysis) hoặc khai phá dữ liệu (data mining).

Với những lợi ích vượt trội như vậy, phân tích thành phần chính (PCA) không chỉ giúp đơn giản hóa và hiểu rõ hơn về dữ liệu phức tạp, mà còn mở ra nhiều cơ hội khai thác thông tin quan trọng từ dữ liệu về phụ nữ tại các tỉnh thành Việt Nam năm 2022. Vì vậy, PCA là lựa chọn ưu tiên trong ba chủ đề phân tích nâng cao được đề xuất.

#1: Cài đặt và load các thư viện cần thiết cho việc phân tích PCA

```r
install.packages("ggplot2")
install.packages("FactoMineR")
install.packages("factoextra")
install.packages("dplyr")
library(ggplot2)
library(FactoMineR)
library(factoextra)
library(dplyr)
```

#2: Chuẩn bị dữ liệu cần thiết cho việc phân tích PCA

```r
load("Women_2022.rda")
data <- Women_2022
# Chuyển đổi cột số
data <- data %>%
  mutate(across(Population_Total:Human_Development_Index, ~ as.numeric(gsub(",", "", .))))

# Tạo cột miền
mien <- c(
  "Ha Noi", "Ha Giang", "Cao Bang", "Bac Kan", "Tuyen Quang", "Lao Cai", "Dien Bien", "Lai Chau",
  "Son La", "Yen Bai", "Hoa Binh", "Thai Nguyen", "Lang Son", "Quang Ninh", "Bac Giang", "Phu Tho",
  "Vinh Phuc", "Bac Ninh", "Hai Duong", "Hai Phong", "Hung Yen", "Thai Binh", "Ha Nam", "Nam Dinh",
  "Ninh Binh"
)
data$Region <- ifelse(data$Province %in% mien, "Bac",
                      ifelse(data$Province %in% c("Thanh Hoa", "Nghe An", "Ha Tinh", "Quang Binh",
                                                  "Quang Tri", "Thua Thien - Hue", "Da Nang", "Quang Nam",
                                                  "Quang Ngai", "Binh Dinh", "Phu Yen", "Khanh Hoa",
                                                  "Ninh Thuan", "Binh Thuan"), "Trung", "Nam"))
```

#3: Thực hiện phân tích PCA cho từng miền

```r
# Miền Bắc :
data_bac <- data %>% filter(Region == "Bac") %>% select(-ID, -Province, -Region)
data_bac_scaled <- scale(data_bac)

pca_bac <- PCA(data_bac_scaled, graph = FALSE)
print(pca_bac)
fviz_eig(pca_bac, addlabels = TRUE, ylim = c(0, 50))
fviz_pca_var(pca_bac, col.var = "contrib", gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"), repel = TRUE)
fviz_pca_ind(pca_bac, col.ind = "cos2", gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"), repel = TRUE)
```

```r
# Miền Trung
data_trung <- data %>% filter(Region == "Trung") %>% select(-ID, -Province, -Region)
data_trung_scaled <- scale(data_trung)

pca_trung <- PCA(data_trung_scaled, graph = FALSE)
print(pca_trung)
fviz_eig(pca_trung, addlabels = TRUE, ylim = c(0, 50))
fviz_pca_var(pca_trung, col.var = "contrib", gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"), repel = TRUE)
fviz_pca_ind(pca_trung, col.ind = "cos2", gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"), repel = TRUE)
```

```r
# Miền Nam
data_nam <- data %>% filter(Region == "Nam") %>% select(-ID, -Province, -Region)
data_nam_scaled <- scale(data_nam)

pca_nam <- PCA(data_nam_scaled, graph = FALSE)
print(pca_nam)
fviz_eig(pca_nam, addlabels = TRUE, ylim = c(0, 50))
fviz_pca_var(pca_nam, col.var = "contrib", gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"), repel = TRUE)
fviz_pca_ind(pca_nam, col.ind = "cos2", gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"), repel = TRUE)
```

#5: Phân tích các thành phần chính cho 3 miền

```r
library(dplyr)
library(FactoMineR)
library(factoextra)

# Hàm xử lý PCA cho từng vùng miền và trả về các thành phần chính
perform_pca_and_extract_pcs <- function(data, region) {
  data_filtered <- data %>%
    filter(Region == region) %>%
    select(-ID, -Province, -Region)

  data_scaled <- scale(data_filtered)

  pca_result <- PCA(data_scaled, graph = FALSE)

  # Lấy các thành phần chính
  pcs <- as.data.frame(pca_result$ind$coord)

  # Gán lại tên cho các thành phần chính
  colnames(pcs) <- paste0("PC", 1:ncol(pcs))

  return(list(pca_result = pca_result, pcs = pcs))
}

# Dữ liệu mẫu của bạn, thay thế bằng dữ liệu thực tế
data <- data.frame(
  ID = 1:100,
  Province = rep(c("Province1", "Province2"), 50),
  Region = rep(c("Bac", "Trung", "Nam"), length.out = 100),
  Var1 = rnorm(100),
  Var2 = rnorm(100),
  Var3 = rnorm(100)
)

# Thực hiện PCA và lấy các thành phần chính cho từng vùng
bac_results <- perform_pca_and_extract_pcs(data, "Bac")
trung_results <- perform_pca_and_extract_pcs(data, "Trung")
nam_results <- perform_pca_and_extract_pcs(data, "Nam")

# Kiểm tra kết quả PCA và các thành phần chính
print(bac_results$pca_result)
print(head(bac_results$pcs))

print(trung_results$pca_result)
print(head(trung_results$pcs))

print(nam_results$pca_result)
print(head(nam_results$pcs))

# Trực quan hóa kết quả PCA
fviz_eig(bac_results$pca_result, addlabels = TRUE, ylim = c(0, 50))
fviz_pca_var(bac_results$pca_result, col.var = "contrib",
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE)
fviz_pca_ind(bac_results$pca_result, col.ind = "cos2",
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE)

fviz_eig(trung_results$pca_result, addlabels = TRUE, ylim = c(0, 50))
fviz_pca_var(trung_results$pca_result, col.var = "contrib",
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE)
fviz_pca_ind(trung_results$pca_result, col.ind = "cos2",
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE)

fviz_eig(nam_results$pca_result, addlabels = TRUE, ylim = c(0, 50))
fviz_pca_var(nam_results$pca_result, col.var = "contrib",
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE)
fviz_pca_ind(nam_results$pca_result, col.ind = "cos2",
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE)
```
