3. Thống kê mô tả:
- Các đại lượng thống kê cơ bản: Tần suất (Frequency), Trung bình(Mean), Trung vị(Median), Phương sai(Variance) và Độ lệch chuẩn(Standard Deviation), Phân phối(Distribution).
- Biểu đồ tóm tắt dữ liệu:
+ Biểu đồ cột (Column Chart): Biểu đồ cột thường được sử dụng để so sánh giữa các giá trị của các biến khác nhau. Trên trục x là các biến và trục y là giá trị của các biến đó.

+ Biểu đồ đường (Line Chart): Biểu đồ đường thường được sử dụng để minh họa sự thay đổi của một biến qua thời gian. Các điểm dữ liệu được nối với nhau bằng các đường thẳng.

+ Biểu đồ hình nến (Candlestick Chart): Biểu đồ hình nến cung cấp thông tin về giá mở cửa, giá cao nhất, giá thấp nhất và giá đóng cửa của một chu kỳ giao dịch. Các hình nến thường được sắp xếp trên một trục thời gian.

+ Biểu đồ điểm và hộp (Scatter and Box Plot): Biểu đồ điểm thường được sử dụng để hiển thị mối quan hệ giữa hai biến. Trong khi đó, biểu đồ hộp thường được sử dụng để mô tả phân phối của một biến dựa trên giá trị trung vị, phân vị và các ngoại lệ.

+ Biểu đồ histogram (Histogram): Biểu đồ histogram được sử dụng để biểu diễn phân phối của một biến liên tục. Dữ liệu được chia thành các nhóm rời rạc và số lượng quan sát trong mỗi nhóm được đếm và biểu diễn trên trục y.

4. Phân tích mối liên hệ giữa các biến:
- Khối lượng giao dịch (Volume): Đây là số lượng cổ phiếu hoặc chứng khoán được giao dịch (mua và bán) trong một khoảng thời gian cụ thể. Biến này đại diện cho mức độ hoạt động giao dịch trên thị trường.
+ Giá cao nhất (High Price): Đây là mức giá cao nhất mà một cổ phiếu hoặc chứng khoán đã đạt được trong một khoảng thời gian cụ thể.
+ Giá thấp nhất (Low Price): Đây là mức giá thấp nhất mà một cổ phiếu hoặc chứng khoán đã đạt được trong một khoảng thời gian cụ thể.
+ Giá mở cửa (Opening Price): Đây là mức giá mà tại đó một cổ phiếu hoặc chứng khoán bắt đầu giao dịch khi thị trường mở cửa.
+ Giá đóng cửa (Closing Price): Đây là mức giá cuối cùng mà tại đó một cổ phiếu hoặc chứng khoán được giao dịch khi thị trường đóng cửa.
- Mối liên hệ giữa các biến khối lượng, giá cao nhất, giá thấp nhất, giá mở cửa và giá đóng cửa được phân tích như sau:
+ Tương quan giữa giá mở cửa và giá đóng cửa: Nếu giá mở cửa và giá đóng cửa có tương quan dương, điều này cho thấy giá thường kết thúc ngày giao dịch gần mức giá bắt đầu ngày giao dịch.
+ Tương quan giữa giá cao nhất và giá thấp nhất: Sự chênh lệch giữa giá cao nhất và giá thấp nhất trong ngày cho biết mức độ biến động của cổ phiếu trong ngày đó.
+ Tương quan giữa khối lượng giao dịch và giá cả: Khối lượng giao dịch có thể có tương quan với các biến giá cả. Ví dụ, tăng khối lượng giao dịch có thể đi kèm với sự tăng giá cổ phiếu do tăng nhu cầu.
+ Mối quan hệ không tuyến tính hoặc không đồng đều: Các mối quan hệ này cung cấp thông tin quan trọng để hiểu và dự báo xu hướng giá cổ phiếu, cũng như phân tích sự biến động và khối lượng giao dịch.
- Ma trận tương quan:
+ Ma trận tương quan là một bảng chứa tất cả các hệ số tương quan giữa các cặp biến. Trong trường hợp này, bạn có thể xây dựng ma trận tương quan cho các biến khối lượng giao dịch, giá cao nhất, giá thấp nhất, giá mở cửa và giá đóng cửa.
+ Mỗi ô trong ma trận sẽ chứa hệ số tương quan giữa hai biến. Hệ số tương quan có thể nằm trong khoảng từ -1 đến 1, với -1 cho biết mối quan hệ nghịch biến hoàn toàn, 1 cho biết mối quan hệ đồng biến hoàn toàn và 0 cho biết không có mối quan hệ tuyến tính nào.
- Hệ số tương quan và độ đồng biến:
+ Hệ số tương quan giữa các biến cung cấp thông tin về mức độ mối quan hệ tuyến tính giữa chúng: Nếu hệ số tương quan gần 1 hoặc bằng -1, nghĩa là mối quan hệ giữa các biến là mạnh mẽ.
+ Độ đồng biến cho biết hướng của mối quan hệ: Nếu độ đồng biến dương, khi một biến tăng, biến kia cũng tăng; nếu độ đồng biến âm, khi một biến tăng, biến kia giảm.
- Biểu đồ Scatter Plot:
+ Biểu đồ scatter plot là một biểu đồ hai chiều trong đó mỗi điểm biểu diễn một cặp giá trị của hai biến: Trong trường hợp này, chúng ta có thể sử dụng scatter plot để biểu diễn mối quan hệ giữa các cặp biến như giá mở cửa và giá đóng cửa hoặc khối lượng giao dịch và giá đóng cửa.
-Việc phân tích các mối quan hệ này cung cấp thông tin quan trọng để hiểu xu hướng thị trường chứng khoán, dự báo giá cổ phiếu, và tối ưu hóa chiến lược đầu tư.

5. Trực quan hóa dữ liệu:

Biểu đồ này đặc biệt hữu ích khi so sánh dữ liệu giữa các nhóm ngành khác nhau hoặc giữa các thời điểm giao dịch. Biểu đồ đường (Line chart) được sử dụng để theo dõi xu hướng thay đổi của giá cổ phiếu theo thời gian, giúp nhận biết xu hướng tăng, giảm của cổ phiếu trong một khoảng thời gian nhất định. Biểu đồ hộp (Box plot) mô tả phân phối của giá cổ phiếu, cung cấp cái nhìn tổng quan về phân phối dữ liệu, bao gồm giá trị trung bình, phạm vi, phần trăm phân vị, và các điểm dữ liệu ngoại lệ, giúp phát hiện các giá trị bất thường.
```r
#THỐNG KÊ MÔ TẢ VÀ BIỂU ĐỒ
tb_khoi_luong <- mean(data$"Khối lượng")
cat("Trung khối lượng:", tb_khoi_luong, "\n")
TV_cao_nhat <- median(data$"Cao nhất")
cat("Trung vị giá cao nhất:", TV_cao_nhat, "\n")
TV_thap_nhat <- median(data$"Thấp nhất")
cat("Trung vị giá thấp nhất:", TV_thap_nhat, "\n")
PS_khoi_luong <- var(data$"Khối lượng")
cat("Phương sai của khối lượng:", PS_khoi_luong, "\n")
DLC_cao_nhat <- sd(data$"Cao nhất")
cat("Độ lệch chuẩn của giá cao nhất:", DLC_cao_nhat, "\n")
DLC_thap_nhat <- sd(data$"Thấp nhất")
cat("Độ lệch chuẩn của giá cao nhất:", DLC_thap_nhat, "\n")
```

```r
# DỮ LIỆU ĐÃ ĐƯỢC CHUẨN HÓA
data$"Khối lượng_Z" <- (data$"Khối lượng" - mean(data$"Khối lượng")) / sd(data$"Khối lượng")
data$"Cao nhất_Z" <- (data$"Cao nhất" - mean(data$"Cao nhất")) / sd(data$"Cao nhất")
data$"Thấp nhất_Z" <- (data$"Thấp nhất" - mean(data$"Thấp nhất")) / sd(data$"Thấp nhất")
data$"Mở cửa_Z" <- (data$"Mở cửa" - mean(data$"Mở cửa")) / sd(data$"Mở cửa")
data$"Đóng cửa_Z" <- (data$"Đóng cửa" - mean(data$"Đóng cửa")) / sd(data$"Đóng cửa")
data_xts <- data[, c('Khối lượng_Z', 'Cao nhất_Z', 'Thấp nhất_Z', 'Mở cửa_Z', 'Đóng cửa_Z')]
print(data_xts)
```

```r
library(ggplot2)
library(dplyr)
library(zoo)
sma_window <- 20
data <- data %>%
mutate(SMA = zoo::rollmean(`Đóng cửa`, k = sma_window, fill = NA))
data <- data %>%
mutate(CCA = (lag(`Đóng cửa`, default = first(`Đóng cửa`)) + `Đóng cửa`) / 2)
ggplot(data, aes(x = Ngày)) +
  geom_segment(aes(x = Ngày, xend = Ngày, y = `Thấp nhất`, yend = `Cao nhất`, color = `Đóng cửa` > `Mở cửa`, linewidth = 1.5)) +
  geom_line(aes(y = SMA), color = 'black', linewidth = 1) +
  geom_line(aes(y = CCA), color = 'violet', linewidth = 1) +
  scale_color_manual(values = c('red', 'green')) +
  labs(title = "Biểu đồ nến với SMA và CCA", y = "Giá", x = "Ngày") +
  theme(legend.position = "none")
```

```r
#kết quả PCA
features_Z <- data[, c('Khối lượng_Z', 'Cao nhất_Z', 'Thấp nhất_Z', 'Mở cửa_Z', 'Đóng cửa_Z')]
pca_result <- prcomp(features_Z, center = TRUE, scale. = TRUE)
summary(pca_result)
```

```r
# Tỷ lệ phương sai giải thích
pca_var <- pca_result$sdev^2
pca_var_ratio <- pca_var / sum(pca_var)
print(pca_var_ratio)
```

```r
# Chuyển đổi dữ liệu gốc sang không gian các thành phần chính
principal_components <- as.data.frame(pca_result$x)
```

```r
# Vẽ biểu đồ Scree Plot
plot(pca_var_ratio, type = "b", main = "Scree Plot", xlab = "Principal Component", ylab = "Proportion of Variance Explained")
```

```r
# Vẽ biểu đồ scatter plot của hai thành phần chính đầu tiên
library(ggplot2)
final_data <- cbind(data["Ngày"], principal_components)
head(final_data)
ggplot(final_data, aes(x = PC1, y = PC2)) +
  geom_point() +
  labs(title = "PCA of Stock Data", x = "Principal Component 1", y = "Principal Component 2")
```
