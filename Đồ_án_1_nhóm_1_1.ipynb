```r
library(dplyr)
#phân nhánh các nước

#châu á
data_asian <- data %>%
  filter()
data_asian <- data %>%
  mutate(Region = ifelse(data$country %in% data("TurKey","Japan","Indian","Vietnam","Thailand","Indonesia","Korea","South Korea","Hong Kong","Taiwan","Israel","Jordan","Saudi Arabia","United Arab Emirates","Lebanon","Malaysia"), "Asia","other"))
data_asian <- filter(data_asian, Region == "Asia")
#bắc mỹ
data_north_america <- data %>%
  mutate(Region = ifelse(data$country %in% data("United States","Canada","Mexico"), "North America","other"))
data_north_america <- filter(data_north_america, Region == "North America")
#châu âu
data_euro <- data %>%
  mutate(Region = ifelse(data$country %in% data("United Kingdom","France","Belgium","Spain","Russia","Ireland","Italy","Poland","Germany","Netherlands","Sweden","Romania"), "Europe","other"))
data_euro <- filter(data_euro, Region == "Europe")
#châu đại dương
data_oceania <- data %>%
  mutate(Region = ifelse(data$country %in% data("Australia","New Zealand"), "Oceania","other"))
data_oceania <- filter(data_oceania, Region == "Oceania")
#châu phi
data_africa <- data %>%
  mutate(Region = ifelse(data$country %in% data("Nigernia","South Africa","Egypt"), "Africa","other"))
data_africa <- filter(data_africa, Region == "Africa")
#châu Mỹ
data_americas <- data %>%
  mutate(Region = ifelse(data$country %in% data("Brazil","Argentina","Colombia"), "Americas","other"))
data_americas <- filter(data_americas, Region == "Americas")
#gộp data
new_data <- rbind(data_asian,data_euro,data_oceania,data_africa,data_americas,data_north_america)
#in data
new_data


```

```r
#mối tương quan giữa view và revenue
library(ggplot2)
# Biểu đồ cột cho doanh thu
plot_doanhthu <- ggplot(new_data, aes(x = Region, y = revenue, fill = "revenue")) +
  geom_bar(stat = "identity", position = "dodge") +
  ylab("revenue") +
  scale_fill_manual(values = "skyblue")

# Biểu đồ đường cho lượt truy cập
plot_luottruycap <- ggplot(new_data, aes(x = Region, y = view, group = 2)) +
  geom_line(color = "red") +
  ylab("view")

# Kết hợp hai biểu đồ vào cùng một biểu đồ
combined_plot <- plot_doanhthu +
  geom_line(data = new_data, aes(x = Region, y = view * max(new_data$revenue)/max(new_data$view), group = 2, color = "view")) +
  geom_point(data = new_data, aes(x = Region, y = view * new_data$revenue/max(new_data$view), color = "view")) +
  scale_color_manual(values = c("red", "red")) +
  theme_minimal() +
  theme(legend.position = "bottom") +
  ylab("revenue / view")

# Hiển thị biểu đồ kết hợp
print(combined_plot)
```

```r
library(ggplot2)

# Chuyển đổi cột "listed_in" sang dạng factor
data$listed_in <- as.factor(data$listed_in)

# Tính toán số lượng xuất hiện của mỗi thể loại phim
genre_counts <- table(data$listed_in)

# Chuyển đổi kết quả thành data frame
genre_counts_df <- as.data.frame(genre_counts)
colnames(genre_counts_df) <- c("movie genre", "Quantity")

# Sắp xếp theo số lượng giảm dần và chọn 10 thể loại phim phổ biến nhất
top_genres <- genre_counts_df[order(-genre_counts_df$`Quantity`), ][1:10, ]

# Tạo biểu đồ cột ngang
ggplot(top_genres, aes(x = reorder(`movie genre`, -`Quantity`), y = `Quantity`)) +
  geom_bar(stat = "identity", fill = "orange") +
  labs(title = "Top 10 movie genre on NETFLIX", x = "movie genre", y = "Quantity") +
  theme_minimal() +
  coord_flip()
```

```r
#biểu đồ rating của các châu lục
library(ggplot2)

# Lấy các giá trị duy nhất của cột 'region' và 'rating'
region <- unique(new_data$Region)
ratings <- unique(new_data$rating)

# Tạo một hình vẽ với kích thước lớn
fig <- ggplot() +
  theme_void() +
  theme(plot.margin = margin(1, 1, 1, 1, "cm"))

# Vòng lặp để vẽ biểu đồ cho mỗi châu lục
for (i in 1:min(length(region), 5)) {
  # Tạo dataframe con chỉ chứa các hàng có giá trị 'region' là châu lục hiện tại
  frame <- subset(new_data, Region == region[i])

  # Tạo đồ thị countplot
  p <- ggplot(frame, aes(x = rating)) +
    geom_bar(fill = "green") +
    labs(title = paste("Content Produced in", region[i]), y = "Content Produced") +
    theme_minimal()

  # Thêm biểu đồ vào hình vẽ
  fig <- fig + annotation_custom(ggplotGrob(p), xmin = 0, xmax = 1, ymin = 1 - i/5, ymax = 1 - (i - 1)/5)
}

# Hiển thị hình vẽ
print(fig)
```

```r
# Mã hoá cột "duration"
new_data$duration <- ifelse(grepl("Season", new_data$duration) & !grepl("\\d+", new_data$duration), 1,
                            ifelse(grepl("min", new_data$duration) & grepl("\\d+", new_data$duration), 1, 0))
new_data$duration
```

```r
# Tính trung bình của cột "view"
mean_view <- mean(new_data$view)

# Mã hoá cột "view"
new_data$view <- ifelse(new_data$view > mean_view, 1, 0)
new_data$view
```

```r
new_data$revenue
```

```r
# Tính trung bình của cột "revenue"
mean_revenue <- mean(new_data$revenue)

# Mã hoá cột "revenue"
new_data$revenue <- ifelse(new_data$revenue > mean_revenue, 1, 0)
new_data$revenue
```

```r
# Load thư viện
library(glmnet)

# Biến movie về 1 còn tv show về 0
#new_data$type <- ifelse(new_data$type == "movie", 1, 0)

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
set.seed(123)
train_index <- sample(1:nrow(new_data), 0.8 * nrow(new_data))
train_data <- new_data[train_index, ]
test_data <- new_data[-train_index, ]

# Xây dựng mô hình hồi quy logistic
model <- glm(revenue ~ view + duration , data = train_data, family = binomial)
model
# Dự đoán xác suất trên tập kiểm tra
probabilities <- predict(model, new_data = test_data, type = "response")
probabilities

# Chuyển đổi xác suất thành nhãn nhị phân dựa trên ngưỡng 0.5
predicted_classes <- ifelse(probabilities > 0.5, 1, 0)

# Tính toán độ chính xác của mô hình
accuracy <- mean(predicted_classes == test_data$revenue)
cat("Độ chính xác của mô hình:", accuracy, "\n")

```
